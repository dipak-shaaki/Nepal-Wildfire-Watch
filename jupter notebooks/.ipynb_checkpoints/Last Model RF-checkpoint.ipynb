```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
import math, pickle, json

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

```python
# LOAD AND PREPARE DATA
```

```python
CSV_PATH = "D:/Final_sem_/Forest_Fire_Prediction/jupter notebooks/final.csv"  
df = pd.read_csv(CSV_PATH, parse_dates=['acq_date'])
print("Dataset Shape:", df.shape)
display(df.head())

# Extract month and day_of_year from date for training
df['month'] = df['acq_date'].dt.month
df['day_of_year'] = df['acq_date'].dt.dayofyear

# Keep only relevant columns
keep_cols = [
    'temperature', 'humidity', 'wind_speed', 'precipitation', 'elevation',
    'latitude', 'longitude', 'month', 'day_of_year', 'fire_occurred'
]
if 'vpd' in df.columns:
    keep_cols.insert(5, 'vpd')  # insert vpd before latitude
df = df[keep_cols]
```

```python
# Compute Missing VPD Values and Clean Data
```

```python
def calculate_vpd(temp_c, humidity):
    es = 0.6108 * np.exp((17.27 * temp_c) / (temp_c + 237.3))
    ea = es * (humidity / 100.0)
    return es - ea

if 'vpd' not in df.columns or df['vpd'].isna().any():
    df['vpd'] = calculate_vpd(df['temperature'], df['humidity'])

df.fillna(df.median(numeric_only=True), inplace=True)
print("Missing values after cleaning:", df.isna().sum().sum())
```

```python
# Split Data into Train/Test (80/20)
```

```python
def stratified_split(X, y, test_size=0.2, random_state=RANDOM_STATE):
    np.random.seed(random_state)
    idx = np.arange(len(y))
    train_idx, test_idx = [], []
    for cls in np.unique(y):
        cls_idx = idx[y == cls]
        np.random.shuffle(cls_idx)
        n_test = int(len(cls_idx) * test_size)
        test_idx.extend(cls_idx[:n_test])
        train_idx.extend(cls_idx[n_test:])
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]
```

```python
X = df[['temperature', 'humidity', 'wind_speed', 'precipitation', 'elevation',
        'vpd', 'latitude', 'longitude', 'month', 'day_of_year']].values
y = df['fire_occurred'].astype(int).values
X_train, X_test, y_train, y_test = stratified_split(X, y)

print("Train shape:", X_train.shape, "| Test shape:", X_test.shape)
```

```python
#  Define Random Forest 
```

```python
def gini(y):
    if len(y) == 0:
        return 0
    counts = Counter(y)
    n = len(y)
    return 1 - sum((counts[c] / n) ** 2 for c in counts)

def best_split_for_column(X_col, y):
    n = len(y)
    if n <= 1:
        return None, 0, None
    order = np.argsort(X_col)
    Xs, ys = X_col[order], y[order]
    parent = gini(ys)
    best_gain, best_t, best_mask = 0, None, None
    for i in range(n - 1):
        if Xs[i] == Xs[i + 1]:
            continue
        t = (Xs[i] + Xs[i + 1]) / 2
        mask = X_col <= t
        left, right = y[mask], y[~mask]
        gain = parent - (len(left) / n) * gini(left) - (len(right) / n) * gini(right)
        if gain > best_gain:
            best_gain, best_t, best_mask = gain, t, mask
    return best_t, best_gain, best_mask

class TreeNode:
    def __init__(self):
        self.is_leaf = False
        self.pred = None
        self.feature_idx = None
        self.threshold = None
        self.left = None
        self.right = None
```

```python
class DecisionTree:
    def __init__(self, max_depth=6, min_samples_split=6, max_features=None, rng=None):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        self.rng = rng or np.random.RandomState()
        self.feature_importances_ = defaultdict(float)

    def fit(self, X, y):
        self.n_features = X.shape[1]
        if not self.max_features:
            self.max_features = int(np.sqrt(self.n_features))
        self.root = self._build(X, y, 0)

    def _build(self, X, y, depth):
        node = TreeNode()
        n = len(y)
        node.pred = np.mean(y)
        if depth >= self.max_depth or n < self.min_samples_split or gini(y) == 0:
            node.is_leaf = True
            return node
        feat_idx = self.rng.choice(self.n_features, self.max_features, replace=False)
        best_gain = 0
        best_feat, best_thresh, best_mask = None, None, None
        for f in feat_idx:
            t, gain, mask = best_split_for_column(X[:, f], y)
            if gain > best_gain:
                best_gain, best_feat, best_thresh, best_mask = gain, f, t, mask
        if best_feat is None:
            node.is_leaf = True
            return node
        self.feature_importances_[best_feat] += best_gain
        node.feature_idx, node.threshold = best_feat, best_thresh
        node.left = self._build(X[best_mask], y[best_mask], depth + 1)
        node.right = self._build(X[~best_mask], y[~best_mask], depth + 1)
        return node

    def _predict_one(self, node, x):
        if node.is_leaf:
            return node.pred
        if x[node.feature_idx] <= node.threshold:
            return self._predict_one(node.left, x)
        else:
            return self._predict_one(node.right, x)

    def predict_proba(self, X):
        return np.array([self._predict_one(self.root, row) for row in X])
```

```python
class RandomForest:
    def __init__(self, n_estimators=100, max_depth=8, min_samples_split=10, max_features=None, random_state=RANDOM_STATE):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        self.rng = np.random.RandomState(random_state)

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_estimators):
            idxs = self.rng.choice(len(y), len(y), replace=True)
            Xb, yb = X[idxs], y[idxs]
            tree_rng = np.random.RandomState(self.rng.randint(0, 1_000_000))
            tree = DecisionTree(self.max_depth, self.min_samples_split, self.max_features, tree_rng)
            tree.fit(Xb, yb)
            self.trees.append(tree)

    def predict_proba(self, X):
        preds = np.array([t.predict_proba(X) for t in self.trees])
        return np.mean(preds, axis=0)

    def predict(self, X):
        return (self.predict_proba(X) >= 0.5).astype(int)
```

```python
# Train Model
```

```python
rf = RandomForest(n_estimators=100, max_depth=8, min_samples_split=8)
rf.fit(X_train, y_train)
```

```python
# Evaluate Model
```

```python
def confusion(y_true, y_pred):
    tp = ((y_true == 1) & (y_pred == 1)).sum()
    tn = ((y_true == 0) & (y_pred == 0)).sum()
    fp = ((y_true == 0) & (y_pred == 1)).sum()
    fn = ((y_true == 1) & (y_pred == 0)).sum()
    return tp, tn, fp, fn

def metrics(y_true, y_pred):
    tp, tn, fp, fn = confusion(y_true, y_pred)
    acc = (tp + tn) / (tp + tn + fp + fn)
    prec = tp / (tp + fp + 1e-9)
    rec = tp / (tp + fn + 1e-9)
    f1 = 2 * prec * rec / (prec + rec + 1e-9)
    return acc, prec, rec, f1
```

```python
train_pred = rf.predict(X_train)
test_pred = rf.predict(X_test)
acc_tr, prec_tr, rec_tr, f1_tr = metrics(y_train, train_pred)
acc_ts, prec_ts, rec_ts, f1_ts = metrics(y_test, test_pred)
```

```python
print(f"Training Accuracy: {acc_tr:.3f}")

```

```python
#  Manual Prediction (User Input)
# ===========================
def manual_fire_prediction(model):
    print("\nEnter environmental values:")
    temp = float(input("Temperature (°C): "))
    hum = float(input("Humidity (%): "))
    wind = float(input("Wind speed (m/s): "))
    precip = float(input("Precipitation (mm): "))
    elev = float(input("Elevation (m): "))
    lat = float(input("Latitude: "))
    lon = float(input("Longitude: "))

    # Derived features
    vpd = calculate_vpd(temp, hum)
    month = 6  # Dummy month (can be dynamic if needed)
    day_of_year = 150  # Dummy day (can be dynamic if needed)

    x = np.array([[temp, hum, wind, precip, elev, vpd, lat, lon, month, day_of_year]])
    prob = model.predict_proba(x)[0]
    label = " High Risk" if prob > 0.66 else (" Moderate Risk" if prob > 0.33 else " Low Risk")

    print(f"\nPredicted Fire Probability: {prob:.3f}")
    print(f"Classification: {label}")

manual_fire_prediction(rf)
```

```python
from matplotlib import pyplot as plt

def plot_confusion_matrix(y_true, y_pred):
    tp, tn, fp, fn = confusion(y_true, y_pred)
    cm = np.array([[tn, fp],
                   [fn, tp]])
    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(cm, cmap='Blues')
    ax.set_title("Confusion Matrix", fontsize=14)
    ax.set_xticks([0,1])
    ax.set_yticks([0,1])
    ax.set_xticklabels(['Pred 0', 'Pred 1'])
    ax.set_yticklabels(['True 0', 'True 1'])
    for (i,j),val in np.ndenumerate(cm):
        ax.text(j,i,f"{val}",ha='center',va='center',color='black',fontsize=12)
    plt.colorbar(im)
    plt.show()

plot_confusion_matrix(y_test, test_pred)
```

```python
# ===========================
# Feature Importance Visualization
# ===========================
importances = defaultdict(float)
for tree in rf.trees:
    for feat, score in tree.feature_importances_.items():
        importances[feat] += score

for feat in importances:
    importances[feat] /= rf.n_estimators

features = ['temperature','humidity','wind_speed','precipitation','elevation','vpd','latitude','longitude']
scores = [importances[i] for i in range(len(features))]

plt.figure(figsize=(8,5))
plt.barh(features, scores)
plt.xlabel("Importance (avg Gini gain)")
plt.ylabel("Feature")
plt.title("Feature Importance — Random Forest (From Scratch)")
plt.gca().invert_yaxis()
plt.show()
```
