```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("forest_dataset.csv")

# Scatter plot
plt.figure(figsize=(8, 6))
colors = df['fire_risk'].map({0: "green", 1: "red"})

plt.scatter(df["longitude"], df["latitude"], 
            c=colors, s=df["temperature"]*5, alpha=0.6, edgecolors="k")

# Labels
for _, row in df.iterrows():
    plt.text(row["longitude"]+0.05, row["latitude"]+0.05, row["forest_name"], fontsize=8)

plt.title("Major Forests in Nepal - Fire Risk Visualization")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
